## Describing data 
#### Histogram
Let us use the built-in dataset airquality which has “Daily air quality measurements in New York, May to September 1973.”

```r
data(airquality)
str(airquality)
```

 We will use the temperature parameter which has 153 observations in degree Fahrenheit.

```r
Temperature <- airquality$Temp
hist(Temperature)
```

- We can see above that there are 9 cells with equally spaced breaks.
- In this case, the height of a cell is equal to the number of observation falling in that cell.
- We can pass in additional parameters to control the way our plot looks.
- You can read about them in the help section ?hist or help(hist).
- Some of the frequently used ones are, main to give the title, xlab and ylab to provide labels for the axes, xlim and ylim to provide range of the axes, col to define color, etc.
- Additionally, with the argument freq=FALSE we can get the probability distribution instead of the absolute frequencies

```r
# histogram with added parameters

hist(Temperature,
     main="Maximum daily temperature at La Guardia Airport",
     xlab="Temperature in degrees Fahrenheit",
     xlim=c(50,100),
     col="darkmagenta",
     freq=FALSE
    )
```

The hist() function returns a list with 6 components

```r
h <- hist(Temperature)
h
```

We see that an object of class histogram is returned which has:
- breaks - places where the breaks occur,
- counts - the number of observations falling in that cell,
- density - the density of cells,
- mids - the midpoints of cells,
- xname - the x argument name and
- equidist - a logical value indicating if the breaks are equally
spaced or not.
• We can use these values for further processing.

For example, in the following example we use the return values to place the counts on top of each cell using the text() function.

```r
h <- hist(Temperature, ylim = c(0,40))
text(h$mids, h$counts, labels = h$counts, adj = c(0.5,-0.5))
```

- We can also define breakpoints between the cells as a vector.
- This makes it possible to plot a histogram with unequal
intervals.
- In such case, the area of the cell is proportional to the number of observations falling inside that cell.

```r
hist(Temperature,
     main="Maximum daily temperature at La Guardia Airport",
     xlab="Temperature in degrees Fahrenheit",
     xlim=c(50,100),
     col="chocolate",
     border="brown",
     breaks=c(55,60,70,75,80,100))
```

truehist() is from the MASS package and scales these counts to give an estimate of the probability density.

```r
# install.packages(MASS)
# loading library MASS
library(MASS)
# Set up a side-by-side plot array
par(mfrow = c(1,2))
# Create a histogram of counts with hist()
hist(Temperature, main = "hist() plot")
# Create a normalized histogram with truehist()
truehist(Temperature, main = "truehist() plot")
```
